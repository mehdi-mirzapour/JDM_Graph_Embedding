```python
from IPython.display import clear_output

def remove_non_related_rows_for_nodes(df_JDM_r_relation,seedwords):
    df_JDM_r_relation_=df_JDM_r_relation.copy()
    lenght=df_JDM_r_relation_.shape[0]
    for row in range(lenght):
        if (df_JDM_r_relation_.loc[row,"ID"] not in seedwords):
            df_JDM_r_relation_=df_JDM_r_relation_.drop(row,axis=0)
        print(row/lenght,"%")
        clear_output(wait=True)
    df_JDM_r_relation_=df_JDM_r_relation_.reset_index(drop=True)
    return(df_JDM_r_relation_)


def remove_non_related_rows_for_links(df_JDM_r_relation,seedwords):
    df_JDM_r_relation_=df_JDM_r_relation.copy()
    lenght=df_JDM_r_relation_.shape[0]
    for row in range(lenght):
        if (df_JDM_r_relation_.loc[row,"Source"] not in seedwords) or (df_JDM_r_relation_.loc[row,"Destination"] not in seedwords):
            df_JDM_r_relation_=df_JDM_r_relation_.drop(row,axis=0)
        print(row/lenght,"%")
        clear_output(wait=True)
    df_JDM_r_relation_=df_JDM_r_relation_.reset_index(drop=True)
    return(df_JDM_r_relation_)



# df_JDM_r_isa[df_JDM_r_isa["Weight"]>230].to_excel("df_JDM_r_isa_230.xlsx")
# df_JDM_r_hypo[df_JDM_r_hypo["Weight"]>120].to_excel("df_JDM_r_hypo_120.xlsx")
# df_JDM_r_syn[df_JDM_r_syn["Weight"]>84].to_excel("df_JDM_r_syn_84.xlsx")
# df_JDM_r_lieu[df_JDM_r_lieu["Weight"]>60].to_excel("df_JDM_r_lieu_60.xlsx")


df_JDM_nodes_removed=remove_non_related_rows_for_nodes(df_JDM_nodes,seedwords)
df_JDM_r_agent_removed=remove_non_related_rows_for_links(df_JDM_r_agent,seedwords)
df_JDM_r_agent_1_removed=remove_non_related_rows_for_links(df_JDM_r_agent_1,seedwords)
df_JDM_r_lieu_removed=remove_non_related_rows_for_links(df_JDM_r_lieu,seedwords)
df_JDM_r_hypo_removed=remove_non_related_rows_for_links(df_JDM_r_hypo,seedwords)
df_JDM_r_raff_sem_removed=remove_non_related_rows_for_links(df_JDM_r_raff_sem,seedwords)
df_JDM_r_syn_removed=remove_non_related_rows_for_links(df_JDM_r_syn,seedwords)
df_JDM_r_isa_removed=remove_non_related_rows_for_links(df_JDM_r_isa,seedwords)


```

```python
import pandas as pd
import numpy as np
import networkx as nx
import node2vec
from gensim.models import Word2Vec

'''
Pipeline for representational learning for all nodes in a graph.
'''

graph_JDM = nx.read_gexf("graph_JDM.gexf")
graph_JDM_words = node2vec.Graph(graph_JDM, True, 1, 1)
graph_JDM_words.preprocess_transition_probs()
walks = graph_JDM_words.simulate_walks(200, 80)   
```

```python
# walks = [list(map(str, walk)) for walk in walks]
model = Word2Vec(walks, size=20, window=10, min_count=0, sg=1, workers=8, iter=1)
model.save("word_embedding_JDM.emd")
```

```python
import node2vec
from gensim.models import Word2Vec,KeyedVectors
from gensim.test.utils import datapath
import gensim
import pandas as pd

df_JDM_all_relations=pd.read_excel("raw_JDM_dataset_Kevin.xlsx")
df_JDM_all_relations["#_of_relations"]=df_JDM_all_relations[["r_syn","r_isa","r_hypo","r_lieu","r_agent_1","r_agent"]].sum(axis=1)
df_JDM_all_relations=df_JDM_all_relations[df_JDM_all_relations["#_of_relations"]>1].reset_index(drop=True)
df_JDM_all_relations=df_JDM_all_relations.drop("#_of_relations",axis=1)


source_list=["source_feature_"+str(id) for id in range(20)]
destination_list=["destination_feature_"+str(id) for id in range(20)]
for id in range(20):
    df_JDM_all_relations.insert(id+4,source_list[id],"")
for id in range(20):
    df_JDM_all_relations.insert(id+24,destination_list[id],"")   

model = KeyedVectors.load('word_embedding_JDM.emd')
        
for row in range(df_JDM_all_relations.shape[0]):
    source_features=model.wv.__getitem__(df_JDM_all_relations["Source_Name"][row])
    destination_features=model.wv.__getitem__(df_JDM_all_relations["Destination_Name"][row])
    both_features=list(source_features)+list(destination_features)
    
    for feature_index in range(20):
        df_JDM_all_relations.iloc[row,feature_index+4]=both_features[feature_index]
        df_JDM_all_relations.iloc[row,feature_index+24]=both_features[feature_index+20]
        
df_JDM_all_relations.to_excel("cleaned_JDM_dataset.xlsx")
```
